```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Read the dataset
df1 = pd.read_csv("/content/sample_data/index_1.csv")
df2 = pd.read_csv("/content/sample_data/index_2.csv")

# You can merge the data if needed
# df = pd.concat([df1, df2], ignore_index=True)
df = df2  # assuming df2 contains complete data

# --- Basic Info ---
print("\n--- Data Overview ---")
print(df.info())
print("\nMissing Values:\n", df.isnull().sum())

# --- Summary Statistics ---
print("\n--- Summary Statistics ---")
print(df.describe())

# --- Top Coffee Items ---
top_items = df['coffee_name'].value_counts().head(10)
print("\nTop Coffee Items:\n", top_items)

# --- Payment Method vs. Average Spend ---
avg_by_payment = df.groupby('cash_type')['money'].mean()
print("\nAverage Spend by Payment Type:\n", avg_by_payment)

# --- T-test: Card vs Cash Spend ---
cash_values = df[df['cash_type'] == 'cash']['money']
card_values = df[df['cash_type'] == 'card']['money']
ttest_result = stats.ttest_ind(card_values, cash_values, equal_var=False)
print("\nT-test Result (Card vs Cash):", ttest_result)

# --- Time Analysis ---
df['datetime'] = pd.to_datetime(df['datetime'])
df['hour'] = df['datetime'].dt.hour
df['day'] = df['datetime'].dt.day_name()

# Plot Hourly Activity
plt.figure(figsize=(10, 5))
sns.histplot(df['hour'], bins=24, kde=False)
plt.title("Purchases by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Count")
plt.xticks(range(0, 24))
plt.grid(True)
plt.show()

# Plot Popular Items
plt.figure(figsize=(10, 5))
sns.barplot(x=top_items.index, y=top_items.values)
plt.title("Top Coffee Beverages")
plt.xlabel("Beverage")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- Heatmap of Payment Type vs Coffee ---
pivot = pd.crosstab(df['coffee_name'], df['cash_type'])
plt.figure(figsize=(12, 6))
sns.heatmap(pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Coffee Type vs Payment Method")
plt.ylabel("Coffee Name")
plt.xlabel("Payment Type")
plt.show()
```
